```python
from langgraph.graph import StateGraph ,START,END
from typing import TypedDict
```

```python
class BMI_State(TypedDict):
    weightKg: float
    heightM: float
    bmi : float
```

```python
graph =StateGraph(BMI_State)
```

```python
def calc(state : BMI_State)-> BMI_State:
    weight=state['weightKg']
    height=state['heightM']
    Bmi=weight/(height**2)
    state['bmi']=Bmi
    return state
```

```python
graph.add_node('bmiCalc',calc)
graph.add_edge(START,'bmiCalc')
graph.add_edge('bmiCalc',END)

workFlow=graph.compile()
```

```python
initialState={ 
    "weightKg":80,
    "heightM": 1.73
}
finalState=workFlow.invoke(initialState)
print(finalState)
```
