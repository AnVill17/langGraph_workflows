```python
from langgraph.graph import StateGraph ,START,END
from typing import TypedDict
from dotenv import load_dotenv
from langchain_google_genai import ChatGoogleGenerativeAI
```

```python
load_dotenv()
```

```python
model =ChatGoogleGenerativeAI(model="gemini-2.5-flash")
```

```python
class blogS(TypedDict):
    topic : str
    outline: str
    blog: str
    evaluation : int
```

```python
def generateOutline(state : blogS)-> blogS:
    topic = state['topic']
    prompt=f'generate a outline for a proffesnal text blog on {topic}'
    response =  model.invoke(prompt).content
    state['outline']=response
    return state
```

```python
def genBlog(state : blogS)-> blogS:
    topic = state['topic']
    outline =state['outline']
    prompt=f'generate a text blog on topic {topic} with outline {outline}'
    response=model.invoke(prompt).content
    state['blog']=response
    return state
```

```python
def eval(state : blogS)-> blogS:
    outline=state['outline']
    blog=state['blog']
    prompt=f'judge the blog {blog} according to the outlin {outline}'
    response=model.invoke(prompt).content
    state['evaluation']=response
    return state
```

```python
graph = StateGraph(blogS)
```

```python
graph.add_node('genOutline',generateOutline)
graph.add_node('genBlog',genBlog)
graph.add_node('eval',eval)

graph.add_edge(START, 'genOutline')
graph.add_edge('genOutline','genBlog')
graph.add_edge('genBlog','eval')
graph.add_edge('eval',END)

workflow =graph.compile()
```

```python
in_state={
    'topic':'ai world and impact'
}
op_state= workflow.invoke(in_state)
print(op_state)
```
